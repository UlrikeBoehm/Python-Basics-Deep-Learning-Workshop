# <font color='red'> Data Visualization & Basic Statistics of PIMA INDIAN DIABETES DATSET

```python
import pandas as pd

PIMA_INDIAN_DIABETES_DATA_filename="pima-indians-diabetes_kaggle.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data=pd.read_csv(PIMA_INDIAN_DIABETES_DATA_filename,names=names)
```

```python
pd.set_option('display.width', 150)
pd.set_option('precision', 2)
description = data.describe()
print(description)
```

```python
# Class Distribution 

class_counts = data.groupby('class').size()
print(class_counts)
```

```python
# Calculate Correlation between variables

corrcoef= data.corr(method='pearson')
print(corrcoef)
```

```python
# Correlation Heatmap
%matplotlib inline 
from matplotlib import pyplot 
fig = pyplot.figure()

ax=fig.add_subplot(111)
cax = ax.matshow(corrcoef, vmin=-1, vmax=1)
fig.colorbar(cax)
pyplot.show()
```

```python
%matplotlib inline 
from pandas.tools.plotting import scatter_matrix

scatter_matrix(data)
pyplot.show()
```
